### MinMax

```python
scaler_minmax = MinMaxScaler(feature_range=(0,1))

iris_df_X = iris_df.drop("target", axis=1)

iris_df_minmax = scaler_minmax.fit_transform(iris_df_X)
```

```python
iris_df.head()
```

```python
iris_df_minmax = pd.DataFrame(iris_df_minmax, columns=iris_df_X.columns)
iris_df_minmax.head()
```

### Standard

<center>Exemplo da diferença entre o Standart Scaler e o Min Max:</center>


<center>https://sebastianraschka.com/Articles/2014_about_feature_scaling.html</center>

```python
scaler_std = StandardScaler(with_mean=True, with_std=True)

iris_df_std = scaler_std.fit_transform(iris_df_X)
```

```python
iris_df.head()
```

```python
iris_df_std = pd.DataFrame(iris_df_std, columns= iris_df_X.columns)
iris_df_std.head()
```

### Robust



<Center>http://benalexkeen.com/feature-scaling-with-scikit-learn/</center>

```python
iris_df.head()
```

### IQL - Intervalo Inter-Quartilico

```python
Q1 = iris_df["sepal width (cm)"].quantile(0.25)
Q3 = iris_df["sepal width (cm)"].quantile(0.75)
IQR = Q3 - Q1

print(IQR)
```

```python
new_min_value = Q1 - 1.2 * IQR
new_max_value = Q3 + 1.2 * IQR

print(new_min_value, new_max_value)
```

```python
iris_df[(iris_df["sepal width (cm)"] >= new_min_value) & 
        (iris_df["sepal width (cm)"] <= new_max_value)]
```

## Feature Selection

### Mutual Information

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
mutual_info_classif(X["petal length (cm)"].values.reshape(-1, 1),
                       y.values.reshape((-1,1)))
```

```python
info_reg = {}
for var in X.columns:
    info_reg[var] = mutual_info_classif(X[var].values.reshape((-1,1)),
                                            y.values.reshape((-1,1)))[0]
```

```python
info_reg_df = pd.DataFrame(data=list(info_reg.values()), index=info_reg.keys(), 
                        columns=['info']).sort_values(by='info', ascending=False)

info_reg_df.plot(kind='bar')
```

### K-Best (F_Score, Chi2)

```python
from sklearn.feature_selection import SelectKBest, f_classif, chi2
```

```python
selectorK = SelectKBest(score_func=chi2, k=2)
selectorK.fit(X,y)

X_clf = pd.DataFrame(selectorK.transform(X))

scores_clf = dict((X.columns[e], selectorK.scores_[e]) 
                  for e in range(X.shape[1]))
```

```python
scores_clf
```

```python
scores_clf_df = pd.DataFrame(data=list(scores_clf.values()), 
                          index=list(scores_clf.keys()),
                         columns =['info'])
```
